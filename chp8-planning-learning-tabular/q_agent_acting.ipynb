```python
from agents import QAgent
import numpy as np
```

```python
load_existing_action_values = True

action_values_file = "q_agent_10000.npy"
```

```python
# define learning rate and discount factor
alpha = 0.1
gamma = 1

# define random exploration constant and gradual scaling factor
epsilon = 0.1
epsilon_scaling = 0.95

# define number of episodes
num_episodes = 100

# initialize agent
q_agent = QAgent(alpha, gamma, epsilon, epsilon_scaling, num_episodes)

# load existing action values if required
if load_existing_action_values:
    action_values = np.load("action_values\\"+action_values_file)
    q_agent.action_values = action_values
    print("Successfully loaded existing action values!")
    q_agent.print_action_values()
    
# release the beast!
q_agent.act()
```
