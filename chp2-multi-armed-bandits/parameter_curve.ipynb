```python
from multi_armed_bandit import BanditAgent
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
num_bandits = 10
steps = 10000

epsilons = [1/128, 1/64, 1/32, 1/16, 1/8, 1/4]
ave_rewards = []

for epsilon in epsilons:

    # agent = BanditAgent(num_bandits, steps, epsilon)
    agent = BanditAgent(num_bandits, steps, epsilon, alpha=0.1)
    agent.play()

    ave_reward = agent.ave_rewards[-1]
    ave_rewards.append(ave_reward)

plt.plot(epsilons, ave_rewards)
plt.xscale("log")
plt.show()
```
