```python
from multi_armed_bandit import BanditAgent
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
num_bandits = 10
steps = 10000
epsilon = 0.2

# agent = BanditAgent(num_bandits, steps, epsilon)
agent = BanditAgent(num_bandits, steps, epsilon, alpha=0.1)
agent.play()

ave_rewards_traj = agent.ave_rewards
optimal_choice_traj = agent.percent_optimal_choice

plt.subplot(1, 2, 1)
plt.plot(ave_rewards_traj)
plt.ylim((0, 5))
plt.subplot(1, 2, 2)
plt.plot(optimal_choice_traj)
plt.ylim((0, 1))
plt.show()
```

```python
num_bandits = 10
steps = 1000

careful_agent = BanditAgent(num_bandits, steps, epsilon=0.001, alpha=0.1)
careful_agent.play()
careful_choice_traj = careful_agent.percent_optimal_choice

neutral_agent = BanditAgent(num_bandits, steps, epsilon=0.1, alpha=0.1)
neutral_agent.play()
neutral_choice_traj = neutral_agent.percent_optimal_choice

crazy_agent = BanditAgent(num_bandits, steps, epsilon=0.3, alpha=0.1)
crazy_agent.play()
crazy_choice_traj = crazy_agent.percent_optimal_choice

plt.plot(careful_choice_traj, label="careful agent")
plt.plot(neutral_choice_traj, label="neutral_agent")
plt.plot(crazy_choice_traj, label="crazy_agent")
plt.ylim((0, 1))
plt.legend()
plt.show()
```
